```python
def calculator():
  def add(n1,n2):
    return n1+n2

  def subtract(n1,n2):
    return n1-n2

  def multiply(n1,n2):
    return n1*n2

  def divide(n1,n2):
    if n2 == 0:
      raise ValueError("Cannot divide by zero.")
    return float(n1)/float(n2)
  def expo(n1,n2):
    return n1**n2

  while True:

    try:
      num1=int(input("\rEnter first number: "))
      num2=int(input("Enter second number: "))
    except ValueError as e:
       print(f"ERROR {e} Please enter valid numbers!")

    print("\nEnter the operation you wish to perform")
    print("\n 1. '+' for Addition\n 2. '-' for Subtraction\n 3. '*' for Multiplication\n 4. '/' for Division\n '^' for Exponential\n 5. Exit")

    ch=input("\n").lower()

    if ch=='+':
      result=add(num1,num2)
      print(f"{num1} + {num2} = {result}")
    elif ch=='-':
      result=subtract(num1,num2)
      print(f"{num1} - {num2} = {result}")
    elif ch=='*':
      result=multiply(num1,num2)
      print(f"{num1} * {num2} = {result}")
    elif ch=='/':
      result=divide(num1,num2)
      print(f"{num1} / {num2} = {result}")
    elif ch=='^':
      result=exp(num1,num2)
      print(f"{num1} / {num2} = {result}")
    elif ch=='exit' or ch=='e':
      print("Exiting.... Byeeeee!")
      break
    else:
      print("Enter valid operator! (Only +,-,*,/ or exit if you wish to exit!)")


print("Welcome to the Hiss-Hiss Calculator")
calculator()



```
